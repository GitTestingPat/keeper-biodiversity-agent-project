# Biodiversity Agent Example
This notebook demonstrates the core Pydantic AI agent functionality without the FastAPI standard.

```python
!pip install 'pydantic-ai-slim' python-dotenv
import os
from pydantic_ai import Agent, RunContext

# Ensure you have set your API key beforehand in your environment
# os.environ['GEMINI_API_KEY'] = 'your_key'

agent = Agent(
    'gemini-2.0-flash',
    system_prompt='You are a specialized Biodiversity Agent. Your goal is to help users understand and protect the environment. You have access to tools that can analyze satellite data, wildlife camera traps, and ocean sensors to detect illegal activities.'
)

@agent.tool
async def analyze_satellite_imagery(ctx: RunContext[None], coordinates: str) -> str:
    print(f"Analyzing imagery at {coordinates}")
    return "ALERT: Satellite imagery indicates a 15% increase in illegal logging."

async def run_chat():
    result = await agent.run(
        'Check the satellite imagery for the Amazon at Lat: -3.4, Lon: -62.1'
    )
    print("\n=== Agent Response ===")
    print(result.data)

# In standard python use asyncio.run(run_chat()), inside Jupyter we can directly await
await run_chat()
```
